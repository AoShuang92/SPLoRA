```python
import matplotlib.pyplot as plt
import numpy as np

# Define LoRA layers
num_layers = 32
layers = np.arange(num_layers)

# Define modification layers for each method
method_1_layers = [0, 1, 2, 3, 4, 27, 28, 29, 30, 31]  # First & last few layers
method_2_layers = [8, 10, 13, 16, 18,19, 24, 25, 26, 30]  # Middle layers

# Assign Y values for visualization (so they donâ€™t overlap)
method_1_y = np.ones(len(method_1_layers)) * 1  # Place at y=1
method_2_y = np.ones(len(method_2_layers)) * 2  # Place at y=2

# Plot
plt.figure(figsize=(10, 3))
plt.scatter(method_1_layers, method_1_y, label="SafeLoRA", color="blue", s=100, marker="o")
plt.scatter(method_2_layers, method_2_y, label="SPLoRA (Ours)", color="red", s=100, marker="x")

# X-axis formatting
plt.xticks(ticks=np.arange(0, num_layers, 5), labels=[f"L{i}" for i in range(0, num_layers, 5)], fontsize=16)
plt.xlim(-1, num_layers)  # Keep limits within range
plt.xlabel("Transformer Blocks", fontsize=16)

# Y-axis formatting (hide ticks)
plt.yticks([])
plt.ylabel("")

# Title & legend
plt.title("Distribution of LoRA Modifications Across Blocks", fontsize=18, fontweight='bold')
plt.legend(fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.6)  # Add grid for clarity
plt.savefig("plot_dist.pdf", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define LoRA layers
num_layers = 32
layers = np.arange(num_layers)

# Define modification layers for each method
method_1_layers = [0, 1, 2, 3, 4, 27, 28, 29, 30, 31]  # First & last few layers
method_2_layers = [8, 10, 13, 16, 18, 19, 24, 25, 26, 30]  # Middle layers

# Assign Y values for visualization (so they donâ€™t overlap)
method_1_y = np.ones(len(method_1_layers)) * 1  # Place at y=1
method_2_y = np.ones(len(method_2_layers)) * 2  # Place at y=2

# Plot
plt.figure(figsize=(10, 3))
plt.scatter(method_1_layers, method_1_y, label="SafeLoRA", color="blue", s=100, marker="o")
plt.scatter(method_2_layers, method_2_y, label="SPLoRA (Ours)", color="red", s=100, marker="x")

# X-axis formatting
plt.xticks(ticks=np.arange(0, num_layers, 5), labels=[f"L{i}" for i in range(0, num_layers, 5)], fontsize=16)
plt.xlim(-1, num_layers)  # Keep limits within range
plt.xlabel("Transformer Blocks", fontsize=16)

# Y-axis formatting (hide ticks)
plt.yticks([])
plt.ylabel("")

# Title & legend
plt.title("Distribution of LoRA Modifications Across Blocks", fontsize=18, fontweight='bold')
plt.legend(fontsize=16, loc="upper left", bbox_to_anchor=(0, 0.9))  # Move legend to top-left outside the plot

plt.grid(axis='x', linestyle='--', alpha=0.6)  # Add grid for clarity
plt.savefig("plot_dist.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
